```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test  = train_test_split(cancer_data.data,
                                                      cancer_data.target,
                                                    stratify=cancer_data.target,
                                                    shuffle=True,
                                                    random_state=144)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
y_train[y_train == 0].size
```

```python
y_test[y_test == 1].size
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3)
```

```python
knn.fit(X_train, y_train)
```

```python
print("Test set of prediction {}".format(knn.predict(X_test)))
```

```python
print("Test set accuracy {:.2f}".format(knn.score(X_test,y_test)))
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_minmax_scaled = scaler.transform(X_train)
X_test_minmax_scaled = scaler.transform(X_test)
```

```python
knn.fit(X_train_minmax_scaled, y_train)
print("Test set accuracy: {:.2f}".format(knn.score(X_test_minmax_scaled,y_test)))
print("Test set accuracy: {:.2f}".format(knn.score(X_train_minmax_scaled,y_train)))
```

```python
train_accuracy =[]
test_accuracy =[]

neighbors = range(1,11)
for n in neighbors:
    knn = KNeighborsClassifier(n_neighbors=n)
    knn.fit(X_train_minmax_scaled,y_train)
    train_accuracy.append(knn.score(X_test_minmax_scaled,y_test))
    test_accuracy.append(knn.score(X_train_minmax_scaled,y_train))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.plot(neighbors,train_accuracy,label="train accuracy")
plt.plot(neighbors,test_accuracy,label="test accuracy")
plt.ylabel("accuracy")
plt.xlabel("neighbors")
plt.legend()
```

```python
knn = KNeighborsClassifier(n_neighbors =5)
knn.fit(X_train_minmax_scaled, y_train)
print("Test set accuracy: {:.2f}".format(knn.score(X_test_minmax_scaled,y_test)))
print("Train set accuracy: {:.2f}".format(knn.score(X_train_minmax_scaled,y_train)))
```
