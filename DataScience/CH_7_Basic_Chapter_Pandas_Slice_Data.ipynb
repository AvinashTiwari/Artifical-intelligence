```python
titanic_female = titanic[titanic['Sex'] == 'female']
titanic_female.head()
                         
```

```python
filter1  = titanic['Sex'] == 'female'
filter2 = titanic['Embarked'] == 'C'
titanic_female_C = titanic[filter1 &  filter2]
titanic_female_C.head()
```
